```
import requests
import pandas as pd
from pyxlsb import convert_date
import numpy as np
from bs4 import BeautifulSoup
url = requests.get('https://github.com/PisutSukpool/BADS7105-CRM-analytics-and-intelligence/tree/main/Homework%2003/Dataset')
soup = BeautifulSoup(url.content, 'html.parser')
div = soup.findAll("span",{"class": "css-truncate css-truncate-target d-block width-fit"})
file_name = []
for i in div:
  if str(i).find('>ac') == -1 and str(i).find('readme') == -1:
    range = str(i)[str(i).find('>Ac')+1:str(i).find('</a>')]
    file_name.append(range)
  elif str(i).find('>ac') != -1 and str(i).find('readme') == -1:
    range = str(i)[str(i).find('>ac')+1:str(i).find('</a>')]
    file_name.append(range)
  else:
    pass
```

```
cnt=0
df=pd.DataFrame()
for f in file_name:
  cnt+=1
  f= f.replace(' ','%20')
  try:
    df2 = pd.read_excel('https://github.com/PisutSukpool/BADS7105-CRM-analytics-and-intelligence/raw/main/Homework%2003/Dataset/'+f,names=['Time','Activity','Pain','Gain'])
    df2['User']= cnt
    df2['FileName']  = f
  except: 
    df2 = pd.read_excel('https://github.com/PisutSukpool/BADS7105-CRM-analytics-and-intelligence/raw/main/Homework%2003/Dataset/'+f, engine='pyxlsb',names=['Time','Activity','Pain','Gain'])
    df2['User']= cnt
    df2['FileName']  = f
  df = df.append(df2)
df.index = pd.RangeIndex(len(df.index))  
```

```
for i,row in df.iterrows():
  if row['FileName'][-1] == 'b':
    df.loc[df.index == i, ['Time']] = convert_date(row['Time'])  
df.head(60)
```

```
df['Seq_Activities']=df.groupby('User').cumcount()
df['Time'] = df['Time'].apply(str)
df['Time_Cls'] = df['Time']
df.head(60)
for ind , row in df.iterrows():
  row_time=row['Time_Cls'].strip().strip(' ').strip('m').strip('a').strip('p').strip(' ')
  if row_time.count(':') == 2 and len(row_time)== 8:   #08:30:00
    df.loc[df.index == ind, ['Time_Cls']] =  row_time[:row_time.find(':')] + row_time[row_time.find(':'):row_time.find(':',3)]

  elif row_time.count(':') == 2 and len(row_time)!= 8: # 1/1/1900  6:00:00 AM
    df.loc[df.index == ind, ['Time_Cls']] = row_time[row_time.find(':')-2 : row_time.find(':',row_time.find(':')-2+3)]

  elif row_time.count(':') == 1 and len(row_time) == 4:   #8:00
    df.loc[df.index == ind, ['Time_Cls']] = '0'+row_time

  elif row_time.count(':') == 1 and len(row_time) != 4:   #08:00
    df.loc[df.index == ind, ['Time_Cls']] = row_time

  elif row_time.count('.') == 1 and len(row_time[:row_time.find('.')]) == 2 and len(row_time[row_time.find('.')+1:]) == 2:  # 13.15
    df.loc[df.index == ind, ['Time_Cls']] = row_time[:row_time.find('.')]+':'+row_time[row_time.find('.')+1:]

  elif row_time.count('.') == 1 and len(row_time[:row_time.find('.')]) == 1 and len(row_time[row_time.find('.')+1:]) == 2:  # 3.15
    df.loc[df.index == ind, ['Time_Cls']] = '0'+row_time[:row_time.find('.')]+':'+row_time[row_time.find('.')+1:]

  elif row_time.count('.') == 1 and len(row_time[:row_time.find('.')]) == 1 and len(row_time[row_time.find('.')+1:]) == 1:  # 3.1
    df.loc[df.index == ind, ['Time_Cls']] = '0'+row_time[:row_time.find('.')]+':'+row_time[row_time.find('.')+1:]+'0'

  elif row_time.count('.') == 1 and len(row_time[:row_time.find('.')]) == 2 and len(row_time[row_time.find('.')+1:]) == 1:  # 13.1
    df.loc[df.index == ind, ['Time_Cls']] = row_time[:row_time.find('.')]+':'+row_time[row_time.find('.')+1:]+'0'

  elif row_time.count('.') == 0 and row_time.count(':') == 0 and len(row_time)==1: #1 
    df.loc[df.index == ind, ['Time_Cls']] = '0'+row_time+':00'

  elif row_time.count('.') == 0 and row_time.count(':') == 0 and len(row_time)==2: #12
    df.loc[df.index == ind, ['Time_Cls']] = row_time+':00'    

  elif row_time.count('.') == 0 and row_time.count(':') == 0 and len(row_time)==3: #800
    df.loc[df.index == ind, ['Time_Cls']] = '0'+row_time[0]+':'+row_time[1:]  

  elif row_time.count('.') == 0 and row_time.count(':') == 0 and len(row_time)==4: #0630
    df.loc[df.index == ind, ['Time_Cls']] = row_time[:2]+':'+row_time[2:]          
```

```
from pythainlp import word_tokenize # ทำการเรียกตัวตัดคำ
from pythainlp.word_vector import * # ทำการเรียก thai2vec
from sklearn.metrics.pairwise import cosine_similarity  # ใช้หาค่าความคล้ายคลึง
import numpy as np
model=get_model() # ดึง model ของ thai2vec มาเก็บไว้ในตัวแปร model
def sentence_vectorizer(ss,dim=300,use_mean=True): # ประกาศฟังก์ชัน sentence_vectorizer
    s = word_tokenize(ss)
    vec = np.zeros((1,dim))
    for word in s:
        if word in model.wv.index2word:
            vec+= model.wv.word_vec(word)
        else: pass
    if use_mean: vec /= len(s)
    return vec
def sentence_similarity(s1,s2):
    return cosine_similarity(sentence_vectorizer(str(s1)),sentence_vectorizer(str(s2)))
```
